# Print Function

```python
# A simple print function is a function in Python that takes any number of variables ( such as strings, numbers etc) and outputs the result. 
# The Print function has a parentheses and the double quotes denotes string.
# Anything written inside double quotes is treated as plain text or strings and not as executable code.

print ("Psy Psy!")
```

```python
# An empty print function prints out a new line

print ("Psy Psy!")
print ()
print ("Psy Psy!")
```

```python
# Print function can also be used to take a portion to the next line by using backslash n (\n)

print ("Psy Psy! \nPsy Psy!")
```

```python
# Putting a space in front of the \n does not affect the output but putting a space AFTER the \n inserts a space at the beginning of the new line. 

print ("Psy Psy! \n Psy Psy!")
```

```python
# Using commas inside the print function concatenates the arguments

print ("Psy Psy!" , "Pika Pika!" , "Psy Psy!" , "Pika Pika!")
```

```python
# Separate print functions create outputs on different lines.

print("My name is ")
print("Fabi Nahian.")
```

```python
# Print function can also have keyword arguments.

# Keyword Argument: end 

print("My name is ", end="")
print("Fabi Nahian.")

# In this example, the end argument has nothing inside the double quotes. So, this argument adds the both of the outputs of the separate print functions in one sentence. 
```

```python
# Keyword Argument: sep

print("My", "name", "is", "Fabi", "Nahian.", sep="-")
print ()
print("My", "name", "is", "Fabi", "Nahian.", sep="*")
print ()
print("My", "name", "is", "Fabi", "Nahian.", sep=" ^__^ ")

# This one separates its outputted arguments as defined by the sep keyword.
```

```python
# Keyword end and Keyword sep can be used together

print("My", "name", "is", sep=" ... ", end=" ^__^ ")
print("Fabi", "Nahian.", sep=" * ", end="*\n")
```
