# Print Function

### The print() function is a built-in function. It prints/outputs a specified message to the screen/consol window.

```python
# A simple print function is a function in Python that takes any number of variables ( such as strings, numbers etc) and outputs the result. 
# The Print function has a parentheses and the double quotes denotes string.
# Anything written inside double quotes is treated as plain text or strings and not as executable code.

print ("Psy Psy!")
```

```python
# An empty print function prints out a new line

print ("Psy Psy!")
print ()
print ("Psy Psy!")
```

```python
# In Python strings the backslash (\) is a special character which announces that the next character has a different meaning, e.g., \n (the newline character) starts a new output line.

# Print function can also be used to take a portion to the next line by using backslash n (\n)

print ("Psy Psy! \nPsy Psy!")
```

```python
# Putting a space in front of the \n does not affect the output but putting a space AFTER the \n inserts a space at the beginning of the new line. 

print ("Psy Psy! \n Psy Psy!")
```

```python
# Using commas inside the print function concatenates the arguments

print ("Psy Psy!" , "Pika Pika!" , "Psy Psy!" , "Pika Pika!")
```

```python
# Separate print functions create outputs on different lines.

print("My name is ")
print("Fabi Nahian.")
```

```python
# Keyword arguments are the ones whose meaning is not dictated by their location, but by a special word (keyword) used to identify them

# Print function can also have keyword arguments.

# Keyword Argument: end 

print("My name is ", end="")
print("Fabi Nahian.")

# In this example, the end argument has nothing inside the double quotes.
# So, this argument adds the both of the outputs of the separate print functions in one sentence. 
```

```python
# Keyword Argument: sep

print("My", "name", "is", "Fabi", "Nahian.", sep="-")
print () # prints a blank line
print("My", "name", "is", "Fabi", "Nahian.", sep="*")
print () # prints a blank line
print("My", "name", "is", "Fabi", "Nahian.", sep=" ^__^ ")

# This one separates its output arguments as defined by the sep keyword.
```

```python
# Keyword 'end' and Keyword 'sep' can be used together

print("My", "name", "is", sep=" ... ", end=" ^__^ ")
print("Fabi", "Nahian.", sep=" * ", end="*\n")
```

# Python Literals

### Literals in Python is defined as the raw data assigned to variables or constants while programming. 
### We mainly have five types of literals which includes string literals, numeric literals, boolean literals, literal collections and a special literal None.




```python
# A literal is data whose values are determined by the literal itself. 
# For example, 123 is a literal but 'c' is not because 'c' as it can be the constant of integration or even the length of a hypotenuse in the sense of a Pythagorean theorem. There are many possibilities.

print("100")  # a string
print(90)    # an integer number
```
